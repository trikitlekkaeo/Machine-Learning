```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#y=ax+b

x=np.linspace(-5,5,100) #สร้าง Array 1 มิติ ตั้งแต่ -5 ถึง 5 จำนวน 100 ชุด
y = 2*x+1
print(y)

plt.plot(x,y,'-r',label = 'y=2x+1')
```

```python
plt.plot(x,y,'-r',label = 'y=2x+1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper center")
plt.title("Graph Y=2x+1")
plt.grid()
plt.show()
```

```python
#Scatter

#y=ax+c
rng = np.random
a=rng.rand(50)*10
b=2*a+rng.randn(50)

plt.scatter(a,b)
plt.xlabel('a')
plt.ylabel('b')
plt.show
```

```python
#Predict value

#Linear Regression Model
model = LinearRegression()

a_new=a.reshape(-1,1) #ทำเป็น Array


#train
model.fit(a_new,b) 

#print(model.score(a_new,b))

#Test Model
afit=np.linspace(-1,11)
afit_new=afit.reshape(-1,1)

bfit = model.predict(afit_new)

#print(afit_new.shape)

#analysis model
plt.scatter(a,b)
plt.plot(afit,bfit)
plt.show()



```
