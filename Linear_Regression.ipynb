```python
import numpy as np
import matplotlib.pyplot as plt
#y=ax+b

x=np.linspace(-5,5,100) #สร้าง Array 1 มิติ ตั้งแต่ -5 ถึง 5 จำนวน 100 ชุด

print(x)
```

```python
y = 2*x+1
print(y)
```

```python
plt.plot(x,y,'-r',label = 'y=2x+1')
```

```python
plt.plot(x,y,'-r',label = 'y=2x+1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper center")
plt.title("Graph Y=2x+1")
plt.grid()
plt.show()
```
